```python
def open_file(file_path):
    file = h5py.File(file_path, 'r')
    for key in file.keys():
        raw_data = np.array(file[key])
        
    return raw_data
```

```python
def convert_roadnetwork(graph_path):
    raw_data=h5py.File(graph_path,'r')
    for key in raw_data.keys():
    #print(graph_data[key],key,graph_data[key].name)
        network_data=raw_data[key]
        g_n=np.asarray(network_data)
    return g_n
```

```python
def save_data(data, file_path):
    with open(file_path , 'wb') as f:
        pickle.dump(data,f)
        f.close()
```

```python
def file_path_collection(file_path):
    dirs = os.listdir(file_path) #Reading all files at D:/ThesisData/ANTWERP19-20/training/
    file_pathes=[]
    file_names = []
    for each_file in dirs:
        each_file_path=file_path + each_file
        file_pathes.append(each_file_path)
        file_names.append(each_file)

    return file_pathes, file_names
```

```python
def aggregate_one_day_data(data_path, x, y, width=200):
    raw_data = open_file(data_path)
    volume_data = raw_data[:, y:y+width, x:x+width, 0:8:2]
    new_data = np.sum(np.sum(volume_data, axis=0), axis=-1)
    

    return new_data
```

```python
def preprocess_data(raw_data_path, x, y, period=30, width=200):
    file_pathes, file_names = file_path_collection(raw_data_path)
    monthly_data =[]
    print(len(file_pathes))
    for j in range(0,180,30):
        month_array = np.zeros((width, width))
        for i in range(j, j+period):
            if i <= len(file_pathes)-1:
                new_data = aggregate_one_day_data(file_pathes[i], x, y)
                month_array = month_array + new_data
                del new_data
        average_data = month_array / period
        
        monthly_data.append(average_data)

    return monthly_data
```
